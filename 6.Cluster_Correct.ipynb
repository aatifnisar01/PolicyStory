```python
import os

# Input and output folder paths
input_folder = 'Output/9.Cluster_Summaries'
output_folder = 'Output/10.FINAL_Cluster_Summaries'

# List of allowed files
allowed_files = [
    'Defense_summary.txt',
    'Economic_Policy_summary.txt',
    'Economic_Recovery_summary.txt',
    'Healthcare_summary.txt',
    'Political_summary.txt',
    'Tax_Policy_summary.txt',
    'Infrastructure_summary.txt',
    'Capital_Markets_summary.txt'
]

# Create output folder if it doesn't exist
os.makedirs(output_folder, exist_ok=True)

# Path for the merged output file
merged_file_path = os.path.join(output_folder, 'others_summary.txt')

# Open the merged file for writing
with open(merged_file_path, 'w') as merged_file:
    # Iterate through each subfolder in the input folder
    for year_folder in os.listdir(input_folder):
        year_folder_path = os.path.join(input_folder, year_folder)

        # Check if it is a directory
        if os.path.isdir(year_folder_path):

            # Iterate through files in the year folder
            for file_name in os.listdir(year_folder_path):

                # Check if the file is not in the allowed list
                if file_name.endswith('.txt') and file_name not in allowed_files:
                    file_path = os.path.join(year_folder_path, file_name)

                    # Append the content of the file to the merged file
                    with open(file_path, 'r') as f:
                        merged_file.write(f'\n--- Content from {year_folder}/{file_name} ---\n')
                        merged_file.write(f.read())

print(f'Merging complete. The file is saved at: {merged_file_path}')
```
